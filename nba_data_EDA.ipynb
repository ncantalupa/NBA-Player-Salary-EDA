# NBA Player Data EDA
### This dataset contains many individual stats about NBA players for the 2022-23 NBA season. This includes the player's salary for that season. The data was found on Kaggle and can be downloaded using the link below.

https://www.kaggle.com/datasets/jamiewelsh2/nba-player-salaries-2022-23-season/data

```python
import pandas as pd 

nba_df = pd.read_csv('nba_2022-23_all_stats_with_salary.csv', index_col=0)
print('Number of Rows: ' + str(nba_df.shape[0]))
print(nba_df.head())
nba_df.describe()
```

### Observations
Using the describe and info functions, we see this dataset has data on 467 NBA players from the 2022-23 NBA Season. Some of the important stats to note here are the Salary, Age, GP(Games Played), and VORP(Value Over Replacement Player). VORP is a good measure of a players value to the team when compared to an average 'replacement' player. 

# Looking at the Data
## Salary by Age

```python
import matplotlib.pyplot as plt 

avg_sal_by_age = nba_df.groupby('Age')['Salary'].mean()
avg_sal_by_age.plot(kind='bar')
plt.xlabel('Age')
plt.ylabel('Average Salary ($10M)')
plt.title('Average Salary by Age')
plt.show()
print(nba_df.groupby('Age').size())
```

```python
nba_df.query('Age == 38')
```

### Observations
- As expected in most other fields, the average salary general trends upward when players hit the prime of their careers which is the range of 28-32. There is then a dropoff after that point as play declines. 

- There is a major outlier here at age 38 where he is the only player at his age and is making an impressive $44M. He is widely known as one of the greatest players of all time and is payed to reflect that. 

## Salary and Player Success

```python
plt.clf()
plt.scatter(x=nba_df["Salary"], y=nba_df["VORP"], marker='o', s=3)
plt.xlabel('Salary (10M)')
plt.ylabel('VORP (Player Effectiveness Rating)')
plt.title('2022-23 Player Salary vs Level of Play')

for i, row in nba_df.iterrows():
    if row["VORP"] > 3.5 or row["Salary"] > 40000000:
        plt.annotate(row['Player Name'], (row['Salary'], row['VORP']), textcoords='offset points', xytext=(0,3), ha='center', fontsize = 6)

plt.show()
```

### Observations
As expected the majority of the league is clustered in the bottom lefthand corner where their salaries are under $10 Million and their VORP is approximately 0, meaning that they are playing about average. This is emphasized that the median VORP is about 0.1. 

This however allows us to see the outliers when it comes to effectiveness and salaries. This gives us examples of some players that could be considered overpaid (John Wall, Russell Westbrook, ...), some players that are paid well because they perform well (Stephen Curry, Giannis Antetokounmpo, ...), and some that are underpaid and performing well (Tyrese Haliburton, Ja Morant, ...). 

## Salary and Grouping by Team
I am interested in looking at salaries by team. One this noticed is that some players are listed as player for multiple teams such as Russell Westbrook as seen below:

```python
nba_df.head()
```

For the purpose of this analysis, I am going to add a column to only count them as being on the last team they played for in the season using the following technique. 

```python
team_adj = []
for i, player in nba_df.iterrows():
    if '/' in player['Team']:
        team_adj.append(str(player['Team']).split('/')[-1])
    else:
        team_adj.append(str(player['Team']))
nba_df['Team_adj'] = team_adj
```

We can now use this Team_adj column to analyze results grouped by team. 

```python
plt.clf()
avg_sal_by_team = nba_df.groupby('Team_adj')['Salary'].mean()
avg_sal_by_team = avg_sal_by_team.sort_values(ascending=False)
avg_sal_by_team.plot(kind='bar')
plt.xlabel('Team')
plt.ylabel('Average Salary ($10M)')
plt.title('Average Salary by Team')
plt.show()
```

### Observations
Here we can see the average salary grouped by team. This allows us to see that some of the most expensive teams are the Clippers(LAC), Suns(PHO), and Warriors(GSW). This is interesting as cross referencing the standings for the 2022-23 season (https://www.basketball-reference.com/leagues/NBA_2023_standings.html) we see that those 3 teams finished in 10th, 11th, and 12th, despite having the highest payrolls. 

We can also see some of the teams with the lowest payrolls being the Rockets(HOU), Spurs(SAS), and Pistons(DET). These teams are more expected, as they finished in 28th, 29th, and 30th at the bottom of the league. One could infer that if you dont spend money on great players, you won't play well. But you also then see that throwing a bunch of money at players doesn't always bring you straight to the top. 

## 3 and D Players
A common archetype of NBA players is called the 3&D, which are players that specialize in shooting 3 pointers as well as playing tough defense. The following analysis will show the best 3&D players based on their 3-Point Percentage and their Defensive Win-Share. 

```python
plt.clf()
THRPvsDWS = nba_df[nba_df["3PA"] > 2.0]
plt.scatter(x=THRPvsDWS["3P%"], y=THRPvsDWS["DWS"], marker='o', s=3)
plt.xlabel('3-Point Percentage')
plt.ylabel('Defensive Win Share')
plt.title('2022-23 Player 3P% vs DWS')

for i, row in THRPvsDWS.iterrows():
    if row["3P%"] > 0.43 or row["DWS"] > 3.0:
        plt.annotate(row['Player Name'], (row['3P%'], row['DWS']), textcoords='offset points', xytext=(0,3), ha='center', fontsize = 6)

plt.show()
```

### Observations
This shows many of the players who are effective in either category. There are few guys that were highly elite in both stats, Al Horford being the best mix of the two. However, there are many players that leaned toward one stat while being high level in the other as well like Donovan Mitchell or Tyrese Maxey.
